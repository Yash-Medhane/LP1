```python
import re
from collections import Counter

# Function to display the content of the file
def display_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        print(content)

# Function to remove punctuation and stop words
def remove_punctuation_and_stopwords(file_path):
    stop_words = {"the", "is", "and", "of", "are", "for", "in"}
    result = []
    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = re.sub(r'[^a-zA-Z\s]', '', word)
                if word.lower() not in stop_words:
                    result.append(word)
    print("File after removing punctuation and stopwords:")
    print(" ".join(result))

# Function to perform suffix stripping
def suffix_stripping(input_file):
    suffixes = [
        "ier", "ied", "iage", "iest", "ies", "iful", "ify", "iness", "ness", "ily", "yer", "ying", "ys", "yable", "yful",
        "al", "ly", "ely", "ing", "ed", "es", "ment", "eing", "led", "lex", "ling"
    ]

    result = []
    with open(input_file, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = re.sub(r'[^a-zA-Z\s]', '', word)  # Remove punctuation
                for suffix in suffixes:
                    if word.endswith(suffix):
                        word = word[:-len(suffix)]  # Strip suffix
                        break
                result.append(word)
    print("File after suffix stripping:")
    print(" ".join(result))

# Function to count word frequency
def count_frequency(file_path):
    with open(file_path, 'r') as file:
        content = file.read().lower().split()
        word_count = Counter(content)

    print("Word frequencies:")
    for word, count in word_count.items():
        print(f"{word}: {count}")

# Main program loop
def main():
    while True:
        print("\n1. Display the file")
        print("2. Remove Stop Words")
        print("3. Suffix Stripping")
        print("4. Count Frequency")
        print("Enter your choice:")

        choice = int(input())

        if choice == 1:
            display_file('Input.txt')
        elif choice == 2:
            remove_punctuation_and_stopwords('Input.txt')
        elif choice == 3:
            remove_punctuation_and_stopwords('Input.txt')  # Run this first to strip punctuations & stop words
            suffix_stripping('without_punctuation_and_stopwords.txt')  # Perform suffix stripping on the result
        elif choice == 4:
            remove_punctuation_and_stopwords('Input.txt')
            suffix_stripping('without_punctuation_and_stopwords.txt')
            count_frequency('suffix_stripping2.txt')  # Use suffix-stripped file for frequency count
        else:
            print("Invalid choice. Exiting.")
            break

if __name__ == '__main__':
    main()
```
