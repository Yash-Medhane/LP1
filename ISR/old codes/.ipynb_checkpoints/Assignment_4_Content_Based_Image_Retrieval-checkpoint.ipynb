```python
import cv2
import numpy as np
from PIL import Image
import imagehash
import matplotlib.pyplot as plt

def compare_images(img_path1, img_path2):
    # Load images
    pil1 = Image.open(img_path1).convert("RGB")
    pil2 = Image.open(img_path2).convert("RGB")

    # --- pHash ---
    h1 = imagehash.phash(pil1)
    h2 = imagehash.phash(pil2)
    dist = h1 - h2 # Hamming distance
    phash_sim = 1 - (dist / 64) # normalized similarity
    print(f"pHash Hamming distance: {dist} => normalized similarity: {phash_sim:.3f}")

    # --- ORB ---
    img1 = cv2.cvtColor(np.array(pil1), cv2.COLOR_RGB2GRAY)
    img2 = cv2.cvtColor(np.array(pil2), cv2.COLOR_RGB2GRAY)
    orb = cv2.ORB_create(nfeatures=500)
    kp1, desc1 = orb.detectAndCompute(img1, None)
    kp2, desc2 = orb.detectAndCompute(img2, None)

    if desc1 is None or desc2 is None:
        print("One of the images produced no ORB descriptors.")
        return

    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    matches = bf.match(desc1, desc2)
    matches = sorted(matches, key=lambda m: m.distance)
    distances = [m.distance for m in matches]
    orb_score = len(matches) / (1 + np.mean(distances)) if matches else 0
    print(f"ORB match score (heuristic): {orb_score:.3f}")
    print(f"Top 10 ORB match distances: {distances[:10]}")

    # --- Visualization ---
    draw = cv2.drawMatches(img1, kp1, img2, kp2, matches[:30], None, flags=2)
    plt.figure(figsize=(12,6))
    plt.imshow(draw, cmap='gray')
    plt.title(f"ORB matches between {img_path1} and {img_path2}")
    plt.axis('off')
    plt.show()

# Example usage; replace with your actual filenames
compare_images("lotus.jpg", "rose.jpg")
```

```python
# --- Install / import ---
!pip install -q opencv-python-headless pillow imagehash matplotlib

import os
from pathlib import Path
from PIL import Image, UnidentifiedImageError
import imagehash
import cv2
import numpy as np
import matplotlib.pyplot as plt
import io

# --- Helper: download some sample images if none exist ---
os.makedirs("sample_images", exist_ok=True)
# These are proper image URLs with parameters
!wget -q -O sample_images/cat1.jpg "https://images.unsplash.com/photo-1518791841217-8f162f1e1131?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80"
!wget -q -O sample_images/dog1.jpg "https://images.unsplash.com/photo-1507143550189-fed454f93097?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80"
!wget -q -O sample_images/cat2.jpg "https://images.unsplash.com/photo-1516972810927-80185027ca84?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80"
!wget -q -O sample_images/dog2.jpg "https://images.unsplash.com/photo-1517423440428-a5a00ad493e8?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80"

# --- Load images robustly ---
def load_images_from_folder(folder_path, max_images=None):
    images = {}
    supported = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')
    for i, f in enumerate(sorted(os.listdir(folder_path))):
        if max_images and i >= max_images:
            break
        if not f.lower().endswith(supported):
            continue
        img_path = Path(folder_path) / f
        try:
            pil = Image.open(img_path).convert("RGB")
            images[f] = pil
        except UnidentifiedImageError:
            print(f"Skipping unreadable/corrupt: {f}")
        except Exception as e:
            print(f"Error loading {f}: {e}")
    return images

# --- Feature extraction ---
def compute_phash(pil_image):
    return imagehash.phash(pil_image)

def compute_orb_descriptor(pil_image, nfeatures=500):
    gray = cv2.cvtColor(np.array(pil_image), cv2.COLOR_RGB2GRAY)
    orb = cv2.ORB_create(nfeatures=nfeatures)
    keypoints, descriptor = orb.detectAndCompute(gray, None)
    return keypoints, descriptor

# --- Similarity metrics ---
def hamming_distance(hash1, hash2):
    return (hash1 - hash2) # imagehash defines '-' as Hamming distance

def orb_match_score(desc1, desc2):
    if desc1 is None or desc2 is None:
        return 0.0
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    matches = bf.match(desc1, desc2)
    if not matches:
        return 0.0
    distances = [m.distance for m in matches]
    return len(matches) / (1 + np.mean(distances)) # normalized score

# --- Index builder ---
def build_index(image_dict):
    index = {}
    for name, pil in image_dict.items():
        phash = compute_phash(pil)
        _, orb_desc = compute_orb_descriptor(pil)
        index[name] = {
            "phash": phash,
            "orb_desc": orb_desc,
            "pil": pil
        }
    return index

# --- Query function ---
def query_image(query_pil, index, top_k=5):
    q_hash = compute_phash(query_pil)
    _, q_orb_desc = compute_orb_descriptor(query_pil)
    results = []
    for name, meta in index.items():
        hdist = hamming_distance(q_hash, meta["phash"])
        orb_score = orb_match_score(q_orb_desc, meta["orb_desc"])
        combined = -hdist + orb_score # simple fusion
        results.append((name, combined, hdist, orb_score))
    results.sort(key=lambda x: x[1], reverse=True)
    return results[:top_k]

# --- Display utilities ---
def show_images_grid(images, titles=None, cols=3, figsize=(10,6)):
    rows = (len(images) + cols - 1) // cols
    plt.figure(figsize=figsize)
    for i, (name, img) in enumerate(images.items()):
        plt.subplot(rows, cols, i+1)
        if isinstance(img, Image.Image):
            plt.imshow(img)
        else:
            plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        if titles:
            plt.title(titles[i], fontsize=8)
        else:
            plt.title(name, fontsize=8)
    plt.tight_layout()
    plt.show()

def display_query_results(query_name, query_pil, results, index):
    print(f"Query: {query_name}")
    plt.figure(figsize=(3,3))
    plt.imshow(query_pil)
    plt.axis('off')
    plt.title("Query Image")
    plt.show()

    matched = {}
    titles = []
    for name, combined, hdist, orb_score in results:
        matched[name] = index[name]["pil"]
        titles.append(f"{name}\nphash={hdist}\nORB={orb_score:.2f}")
    show_images_grid(matched, titles=titles, cols=3)

# --- Build and index sample images ---
images = load_images_from_folder("sample_images", max_images=100)
if not images:
    raise RuntimeError("No images loaded; put some images into sample_images/")
index = build_index(images)

# --- Option: upload a query image ---
from google.colab import files
uploaded = files.upload() # choose a query image from local
query_pil = None
query_name = None
for name, data in uploaded.items():
    if len(data) == 0:
        print(f"{name} is empty; skip.")
        continue
    try:
        img = Image.open(io.BytesIO(data))
        img.verify()
        img = Image.open(io.BytesIO(data)).convert("RGB")
        query_pil = img
        query_name = name
        print(f"Using uploaded query: {name}, size={img.size}")
        break
    except Exception as e:
        print(f"Failed to load {name}: {e}")

# Fallback: if no upload, use one from index
if query_pil is None:
    query_name, meta = next(iter(index.items()))
    query_pil = meta["pil"]
    print(f"No valid upload; using sample image '{query_name}' as query.")

# --- Run retrieval ---
results = query_image(query_pil, index, top_k=6)
display_query_results(query_name, query_pil, results, index)
```

```python
# --- Installs & imports ---
!pip install -q opencv-python-headless pillow imagehash matplotlib transformers ftfy accelerate

import os
from pathlib import Path
from PIL import Image, UnidentifiedImageError
import imagehash
import cv2
import numpy as np
import matplotlib.pyplot as plt
import io
import torch
from transformers import CLIPModel, CLIPProcessor
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files

# --- Download sample images if needed ---
os.makedirs("sample_images", exist_ok=True)
!wget -q -O sample_images/cat1.jpg "https://images.unsplash.com/photo-1518791841217-8f162f1e1131?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80"
!wget -q -O sample_images/dog1.jpg "https://images.unsplash.com/photo-1507143550189-fed454f93097?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80"
!wget -q -O sample_images/cat2.jpg "https://images.unsplash.com/photo-1516972810927-80185027ca84?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80"
!wget -q -O sample_images/dog2.jpg "https://images.unsplash.com/photo-1517423440428-a5a00ad493e8?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80"

# --- Image loading ---
def load_images_from_folder(folder_path, max_images=None):
    images = {}
    supported = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')
    for i, f in enumerate(sorted(os.listdir(folder_path))):
        if max_images and i >= max_images:
            break
        if not f.lower().endswith(supported):
            continue
        img_path = Path(folder_path) / f
        try:
            pil = Image.open(img_path).convert("RGB")
            images[f] = pil
        except UnidentifiedImageError:
            print(f"Skipping unreadable/corrupt: {f}")
        except Exception as e:
            print(f"Error loading {f}: {e}")
    return images

# --- Feature extractors ---
def compute_phash(pil_image):
    return imagehash.phash(pil_image)

def compute_orb_descriptor(pil_image, nfeatures=500):
    gray = cv2.cvtColor(np.array(pil_image), cv2.COLOR_RGB2GRAY)
    orb = cv2.ORB_create(nfeatures=nfeatures)
    keypoints, descriptor = orb.detectAndCompute(gray, None)
    return keypoints, descriptor

# CLIP setup
device = "cuda" if torch.cuda.is_available() else "cpu"
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

def get_clip_embedding(pil_image):
    inputs = clip_processor(images=pil_image, return_tensors="pt").to(device)
    with torch.no_grad():
        emb = clip_model.get_image_features(**inputs) # (1, D)
    emb = emb / emb.norm(p=2, dim=-1, keepdim=True) # normalize
    return emb.cpu().numpy() # shape (1, D)

# --- Similarity helpers ---
def hamming_distance(hash1, hash2):
    return (hash1 - hash2) # lower is more similar

def orb_match_score(desc1, desc2):
    if desc1 is None or desc2 is None:
        return 0.0
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    try:
        matches = bf.match(desc1, desc2)
    except:
        return 0.0
    if not matches:
        return 0.0
    distances = [m.distance for m in matches]
    return len(matches) / (1 + np.mean(distances))

def normalize_scores(raw):
    arr = np.array(raw, dtype=float)
    if arr.max() == arr.min():
        return np.ones_like(arr)
    return (arr - arr.min()) / (arr.max() - arr.min())

# --- Index building with CLIP ---
def build_index_with_clip(image_dict):
    index = {}
    for name, pil in image_dict.items():
        phash = compute_phash(pil)
        _, orb_desc = compute_orb_descriptor(pil)
        clip_emb = get_clip_embedding(pil)
        index[name] = {
            "phash": phash,
            "orb_desc": orb_desc,
            "clip_emb": clip_emb,
            "pil": pil
        }
    return index

# --- Fusion query using phash + ORB + CLIP ---
def query_with_clip_fusion(query_pil, index, top_k=5, weights=(0.1, 0.0, 0.9)):
    # weights: (w_phash, w_orb, w_clip) sum roughly to 1
    w_phash, w_orb, w_clip = weights
    q_hash = compute_phash(query_pil)
    _, q_orb_desc = compute_orb_descriptor(query_pil)
    q_clip = get_clip_embedding(query_pil)

    names = []
    phash_dists = []
    orb_scores = []
    clip_sims = []

    for name, meta in index.items():
        hdist = hamming_distance(q_hash, meta["phash"])
        orb_score = orb_match_score(q_orb_desc, meta["orb_desc"])
        clip_sim = cosine_similarity(q_clip, meta["clip_emb"])[0][0]
        names.append(name)
        phash_dists.append(hdist)
        orb_scores.append(orb_score)
        clip_sims.append(clip_sim)

    phash_sim = normalize_scores([-d for d in phash_dists]) # invert distance
    orb_norm = normalize_scores(orb_scores)
    clip_norm = normalize_scores(clip_sims)

    fused = []
    for i, name in enumerate(names):
        combined = w_phash * phash_sim[i] + w_orb * orb_norm[i] + w_clip * clip_norm[i]
        fused.append({
            "name": name,
            "fused_score": combined,
            "phash_dist": phash_dists[i],
            "orb_score": orb_scores[i],
            "clip_sim": clip_sims[i],
            "pil": index[name]["pil"]
        })

    fused.sort(key=lambda x: x["fused_score"], reverse=True)
    return fused[:top_k]

# --- Visualization ---
def show_images_grid(images, titles=None, cols=3, figsize=(10,6)):
    rows = (len(images) + cols - 1) // cols
    plt.figure(figsize=figsize)
    for i, (name, img) in enumerate(images.items()):
        plt.subplot(rows, cols, i+1)
        if isinstance(img, Image.Image):
            plt.imshow(img)
        else:
            plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        if titles:
            plt.title(titles[i], fontsize=8)
        else:
            plt.title(name, fontsize=8)
    plt.tight_layout()
    plt.show()

def display_clip_results(query_name, query_pil, results):
    print(f"Query: {query_name}")
    plt.figure(figsize=(3,3))
    plt.imshow(query_pil)
    plt.axis('off')
    plt.title("Query Image")
    plt.show()
    for rank, item in enumerate(results, 1):
        print(f"{rank}. {item['name']} | fused={item['fused_score']:.3f} | phash_dist={item['phash_dist']} | ORB={item['orb_score']:.2f} | CLIP_sim={item['clip_sim']:.3f}")
    matched = {item['name']: item['pil'] for item in results}
    titles = [f"{it['name']}\nfused={it['fused_score']:.2f}" for it in results]
    show_images_grid(matched, titles=titles, cols=3)

# --- Main execution ---
# Load or prepare gallery
images = load_images_from_folder("sample_images", max_images=200)
if not images:
    raise RuntimeError("No images found in sample_images/")
index = build_index_with_clip(images)

# Upload query image
uploaded = files.upload()
query_pil = None
query_name = None
for name, data in uploaded.items():
    if len(data) == 0:
        print(f"{name} empty, skipping.")
        continue
    try:
        img = Image.open(io.BytesIO(data))
        img.verify()
        img = Image.open(io.BytesIO(data)).convert("RGB")
        query_pil = img
        query_name = name
        print(f"Using uploaded query: {name}, size={img.size}")
        break
    except Exception as e:
        print(f"Failed to load {name}: {e}")

# fallback to sample if no upload
if query_pil is None:
    query_name, meta = next(iter(index.items()))
    query_pil = meta["pil"]
    print(f"No valid upload; using '{query_name}' as query.")

# Retrieval with CLIP-heavy fusion (flower vs animals will separate better)
results = query_with_clip_fusion(query_pil, index, top_k=6, weights=(0.1, 0.0, 0.9))
display_clip_results(query_name, query_pil, results)
```

```python
def display_text_clip_results(query_text, results):
    print(f"Text Query: {query_text}")
    for rank, item in enumerate(results, 1):
        print(f"{rank}. {item['name']} | CLIP_sim={item['clip_sim']:.3f}")
    matched = {item['name']: item['pil'] for item in results}
    titles = [f"{it['name']}\nCLIP={it['clip_sim']:.2f}" for it in results]
    show_images_grid(matched, titles=titles, cols=3)
```
