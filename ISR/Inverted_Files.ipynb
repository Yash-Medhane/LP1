```python
import os, string

folder = "inverted"           # folder with text files
index = {}
```

```python
for file in os.listdir(folder):
    if file.endswith(".txt"):
        text = open(os.path.join(folder, file), encoding="utf-8").read().lower()
        for p in string.punctuation:
            text = text.replace(p, " ")
        for word in set(text.split()):
            index.setdefault(word, []).append(file)
```

```python
print("✅ Inverted Index Created!\n" + "-"*40)
for term, docs in list(index.items())[:10]:
    print(f"{term} → {docs}")
```

```python
print("-"*40)
q = input("Enter word to search: ").lower().strip()
print(index.get(q, f"No documents found for '{q}'"))
```
